```python
# reducimos los registros a analizar a 200.000
df_vuelos_red = df_vuelos.sample(n=200000)
df_vuelos_red.info()
```

### Breve descripción de las columnas

- Unnamed: Indice de los registros.
- Year: Año del registro.	
- Month: Mes del registro en formato numérico.
- DayofMonth: Día del mes del registro en formato numérico.
- DayOfWeek: Día de la semana del registro en formato numérico.
- DepTime: Hora local de salida del vuelo (hhmm).
- CRSDepTime: Hora local prevista de salida del vuelo (hhmm).
- ArrTime: Hora local de llegada del vuelo (hhmm).
- CRSArrTime: Hora local prevista de llegada del vuelo(hhmm).
- UniqueCarrier: Identificación abreviada de la compañia.
- FlightNum: Número de vuelo.
- TailNum: Número de cola del vuelo.
- ActualElapsedTime: Tiempo transcurrido de vuelo en minutos.
- CRSElapsedTime: Tiempo previsto de vuelo en minutos.
- AirTime: Tiempo de vuelo en minutos( en aire).
- ArrDelay: Tiempo de retraso llegada en minutos.
- DepDelay: Tiempo de retraso salida en minutos.
- Origin: Código aeropuerto de origen.
- Dest: Código aeropuerto de destino.
- Distance: Distancia en millas.
- TaxiIn: Rodaje del avión puerta embarque-pista de despegue en minutos.
- TaxiOut: Rodaje del avión pista de despegue-puerta de embarque en minutos.
- Cancelled: Indicador de vuelo cancelado (1= yes, 0= no).
- CancellationCode: Razón de la cancelación (A = carrier, B = weather, C = NAS, D = security).
- Diverted: Indicador de vuelo desviado (1= yes, 0= no).
- CarrierDelay: Retraso causa Carrier en minutos.
- WeatherDelay: Retraso causa Weather en minutos.
- NASDelay: Retraso causa NAS en minutos.
- SecurityDelay: Retraso causa Security en minutos.
- LateAircraftDelay: Retraso causa LateAircraft en minutos.

### Columnas a eliminar y explicación

- Unnamed: 0, es una columna con información repetida.
- Year, Month, DayofMonth las unificamos en una nueva columna llamada 'Date'
- DepTime, CRSDepTime, ArrTime y CRSArrTime se pueden borrar ya que ya tenemos columnas que nos indican el tiempo empleado en los vuelos (ActualElapsedTime y CRSElapsedTime).
- Podemos eliminar también los campos TaxiIn y TaxiOut ya que no son relevantes en el cálculo total del retraso.
- CancellationCode se puede borrar, ya que para los motivos del ejercicio nos basta saber si el vuelo ha sido cancelado, pero no necesitamos la causa.

```python
# unificamos Year, Month, DayofMonth en Date y colocamos esta última al principio de las columnas

df_vuelos_red['Date']=pd.to_datetime(dict(year=df_vuelos['Year'], month=df_vuelos['Month'], day=df_vuelos['DayofMonth']))
df_vuelos_red.insert(1, 'Date', df_vuelos_red.pop('Date'))
```

```python
df_vuelos_red.head()
```

```python
# borramos las columnas que no necesitamos
df_vuelos_red.drop(['Unnamed: 0','Year','Month','DayOfWeek','DayofMonth','DepTime','CRSDepTime','ArrTime','CRSArrTime','CancellationCode','TaxiIn','TaxiOut'],axis=1,inplace=True)
```

```python
df_vuelos_red.head()
```

Para las columnas CarrierDelay, WeatherDelay, NASDelay , SecurityDelay, LateAircraftDelay que no tienen datos, relleno los mismos con el valor 0 para que no interfieran en los futuros cálculos en el caso de ser necesarias. De esta manera el dataset queda limpio.

```python
df_vuelos_red=df_vuelos_red.fillna(0)
df_vuelos_red.isnull().sum()
```

### Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).

```python
# Calculamos la velocidad media del vuelo en mph cómo la división entre la distancia ('Distance') y el tiempo del avión 
# en el aire ('Airtime'). Colocamos la nueva columna junto a la Distancia.


df_vuelos_red['Ave_Speed'] = round(df_vuelos_red['Distance']/ (df_vuelos_red['AirTime'] / 60),2)
df_vuelos_red.insert(12, 'Ave_Speed', df_vuelos_red.pop('Ave_Speed'))
df_vuelos_red.head()
```

```python
# Comprobamos si el avión ha llegado tarde o no en función del valor del campo 'ArrDelay'. 
# Si este es >= 0 consideramos que sí ha llegado tarde. Colocamos la nueva columna junto a la columna "ArrDelay".

df_vuelos_red['Delayed'] = np.where(df_vuelos_red['ArrDelay'] >= 0, 1, 0)
df_vuelos_red.insert(8, 'Delayed', df_vuelos_red.pop('Delayed'))
df_vuelos_red.head()
```

### Fes una taula de les aerolínies amb més endarreriments acumulats.

```python
# Creamos una tabla temporal donde añadimos el campo compañia y el valor de 'ArrDelay'. En esta tabla quitamos aquellos valores
# donde retrasado = 0, ya que 'ArrDelay' tienen valores negativo o cero y los agrupamos por compañia sumando el valor de 'ArrayDelay'.
# Finalmente renombramos las columnas 'Delayed' y 'ArrDelay'
df_vuelos_red_comp = df_vuelos_red.loc[:,['UniqueCarrier','ArrDelay','Delayed']]
df_vuelos_red_comp = df_vuelos_red_comp.drop(df_vuelos_red_comp[df_vuelos_red_comp['Delayed']==0].index).groupby(['UniqueCarrier']).sum().sort_values('ArrDelay', ascending=False)
df_vuelos_red_comp.rename(columns = {'ArrDelay':'AcumDelay'}, inplace = True)
df_vuelos_red_comp.rename(columns = {'Delayed':'NomDelays'}, inplace = True)
df_vuelos_red_comp
```

### Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.

```python
df_vuelos_red[['Origin','Dest','Distance']].sort_values(by=['Distance'],ascending=False).drop_duplicates()
```

```python
df_vuelos_red[['Origin','Dest','ArrDelay']].groupby(['Origin', 'Dest']).sum().reset_index().sort_values(by=['ArrDelay'],ascending=False).query('ArrDelay > 0')
```
